        Fit the model with the data from the features (X): model.fit(X)
        And assign cluster to each sample: p=model.predict(X)
        Find the unique clusters with np.unique() method
        Scatterplot each sample, colored by its cluster

   4. Try with cluster algorithm: DBSCAN

```python
#1
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)

#2
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
#3
from sklearn.cluster import AffinityPropagation
import numpy as np
model = AffinityPropagation(damping=0.7)
model.fit(X)
p=model.predict(X)
np.unique(p)
plt.scatter(X[:,0], X[:,1], c=p)
plt.show()
```

```python
#4
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=.1, min_samples=2).fit(X)
labels = clustering.labels_
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```

```python
#5
from sklearn.cluster import MeanShift
clustering = MeanShift(bandwidth=1).fit(X)
labels = clustering.labels_
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```

```python
#6
from sklearn.cluster import Birch
brc = Birch(n_clusters=None)
brc.fit(X)
b = brc.predict(X)
plt.scatter(X[:,0], X[:,1], c=b)
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=2, random_state=0).fit(X)
gm.means_
g = gm.predict(X)
plt.scatter(X[:,0], X[:,1], c=g)
plt.show()
```
