### Part 1
1. Get data from: "https://en.wikisource.org/wiki/Portal:State_of_the_Union_Speeches_by_United_States_Presidents"
2. Using BeatifullSoup get all the speeches from 1900-2022
3. Load all speech urls into a dictionary with year as key. Hint (get year with regex: `r"\b(19|20)\d{2}\b"`)
4. Loop through dictionary and save content of each speech in [year].txt files

```python
import requests as req
import re
from bs4 import BeautifulSoup
url = "https://en.wikisource.org/wiki/Portal:State_of_the_Union_Speeches_by_United_States_Presidents"
page =  req.get(url).text
soup = BeautifulSoup(page, "html.parser")
print(soup.text[:100])
```

```python
element = soup.find_all('li')[144:269]
element
```

```python
speech_url = []
speech_year = []
for link in element:
    speech_url.append("https://en.wikisource.org"+link.a.get('href'))
    years = re.compile(r"\b(19|20)\d{2}\b")
    spy = years.search(link.text)
    speech_year.append(spy.group())
speech_dict = dict(zip(speech_year,speech_url))
speech_dict
```

```python
for year in speech_dict:
    speech_page = req.get(speech_dict[year]).text
    soup = BeautifulSoup(speech_page, "html.parser")
    speech = soup.select("div > p")[1:-1]
    
    with open("../my_notebooks/speeches/"+year+".txt", 'w+') as f:
        for s in speech:
            f.write(s.text)
    
    
```

4. From the developed text files in part 1, create a dictionary with year:speech


```python
year_speech_dict = {}

for year in speech_year:
    with open('../my_notebooks/speeches/'+year+'.txt') as f:
        year_speech_dict[year] = f.read()
        
#year_speech_dict
```

5. Clean text by change all to lowercase and remove '\n'


```python
for year in year_speech_dict:
    year_speech_dict[year] = re.sub('\n', '', year_speech_dict[year].lower())
    
#year_speech_dict
```

6. Get words from texts (from nltk.tokenize import word_tokenize). Clean text by removing stop words (from nltk.corpus import stopwords) and all non-alphabetic characters (including , and .)


```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

speech_tokens = []

for year in year_speech_dict:
    tokens = word_tokenize(year_speech_dict[year])
    for word in tokens:
        speech_tokens.append(word)
stop_words = set(stopwords.words('english'))
speech_tokens = [w for w in speech_tokens if not w in stop_words]
        
speech_tokens[:20]
```

```python
cleaned_tokens = []
for token in speech_tokens:
    if token.isalpha():
        cleaned_tokens.append(token)
cleaned_tokens[:20]
```

7. Use from nltk.stem import WordNetLemmatizer to lemmatize all texts

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemmatized_words=[lemmatizer.lemmatize(word=word,pos='v') for word in cleaned_tokens]
lemmatizeddf= pd.DataFrame({'original_word': cleaned_tokens,'lemmatized_word': lemmatized_words})
lemmatizeddf=lemmatizeddf[['original_word','lemmatized_word']]
lemmatizeddf[:20]
```
