```python
import bs4
import requests

#1
r = requests.get('https://en.wikipedia.org/wiki/Alan_Turing')
soup = bs4.BeautifulSoup(r.text, 'html.parser')
print(soup.prettify()[:1500])


```

## 2

```python
title = soup.select('Title')
print(title)
```

## 3

```python
ptag = soup.select('p')
print(ptag[1].text)
```

## 4

```python
toc = soup.select('div[id=toc]')

for t in toc:
    print(t.getText())
```

## 5

```python
toc = soup.select('div[id=toc]>ul>li>ul>li>a')
my_dict =  {}
for t in toc:
    my_dict[t.getText()] = t.get('href')
my_dict
```

## 6
 Create a function that can take a word and look for it in the dictionary keys and then return the content from the first link that matches. Return the next p elements until next headline (h3 element).

```python
def myfunction(word):
    result = ''
    for key, value in my_dict.items():
        if word in key:
            for tag in soup.find('span', {'id': value[1:]}).find_all_next():
                if tag.name == 'p': result += tag.getText()
                elif tag.name == 'h3':break
    print(result)
    
myfunction('1.1 Family')
    
    
```

## 1
1. Using regex find out how many times 'Turing' is used in the article


```python
import re
find_turing = re.compile(r'Turing')
found = find_turing.findall(soup.getText())
print(len(found))
```

## 2
2. Regex: Find all the sentences that has a year in them (sentense defined by: starting at \n or dot or comma and ending at dot or comma.

```python
find_turing = re.compile(r'^\\n+|.+\d{4,4}.')
sen = find_turing.findall(soup.getText())

for s in sen:
    print(s)
```
